2. Configuration Space Exploration

```python

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle, Polygon
import numpy as np

# Define polygonal and circular obstacles
# బహుభుజ మరియు వృత్తాకార ఆటంకాలను నిర్వచించు
polygon1 = np.array([[3, 3], [4, 5], [2, 6]])
polygon2 = np.array([[6, 1], [7, 2], [5.5, 3]])
circles = [((7, 7), 0.8), ((2, 8), 0.5)]
robot_radius = 0.5  # రోబో యొక్క వ్యాసార్థం

# Inflate circles (robot size ని జోడించు)
# వృత్తాల వ్యాసార్థాన్ని పెంచడం
def inflate_circles(circles, r):
    return [(center, radius + r) for (center, radius) in circles]

# ప్లాట్ మొదలుపెట్టు
fig, ax = plt.subplots()

# అసలు బహుభుజాలు గీయు
for i, poly in enumerate([polygon1, polygon2]):
    ax.add_patch(Polygon(poly, color='gray', label='Polygon Obstacle' if i == 0 else ""))

# అసలు వృత్తాలు (ప్రత్యేక రంగులతో)
for i, (center, radius) in enumerate(circles):
    unique_color = 'purple' if i == 0 else 'green'  # వేరే వేరే రంగులు
    ax.add_patch(Circle(center, radius, color=unique_color, alpha=0.6, label='Circle Obstacle' if i == 0 else ""))

# పెద్దవిగా చేసిన వృత్తాలు (ఇన్‌ఫ్లేటెడ్)
inflated = inflate_circles(circles, robot_radius)
for i, (center, radius) in enumerate(inflated):
    unique_color = 'violet' if i == 0 else 'limegreen'  # వేరే రంగులు
    ax.add_patch(Circle(center, radius, color=unique_color, alpha=0.3, label='Inflated Circle' if i == 0 else ""))

# ప్లాట్ పరిమితులు
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# చివరి సెట్టింగులు
ax.set_aspect('equal')  # సమాన స్కేలు
ax.grid(True)
ax.set_title("Configuration Space with Polygons and Unique Circles")  # శీర్షిక
ax.legend(loc='lower left')  # లెజెండ్ దిగువవైపు
plt.show()
```
